```python
import numpy as np
import matplotlib.pyplot as plt

from torchvision.datasets import MNIST
from torchvision.transforms import ToTensor

from sklearn.mixture import GaussianMixture
```

```python
datapath = 'data'

data_test = MNIST(
    root = datapath, 
    train = False, 
    transform = ToTensor(),
)
```

```python
digit = 4
x_data = [image.numpy() for image, label in data_test if label==digit]
x_data = np.asarray(x_data).reshape(-1, 28*28)
```

```python
x_data.shape
```

```python
gmm = GaussianMixture(n_components=10, random_state=0, max_iter=100).fit(x_data)
```

```python
x_sample, y_sample = gmm.sample(10)
```

```python
x_sample[x_sample < 0] = 0
images_sample = x_sample.reshape(-1, 28, 28)

for image in images_sample:
    plt.figure()
    plt.imshow(image, cmap='binary')
    plt.draw()
```
